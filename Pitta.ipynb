Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read Dataset

```python
data=pd.read_excel("./Dataset/Flow_history.xls")
data=pd.DataFrame(np.array(data),columns=['Year','Flow','Height'])
print(data.describe())

data_year= data["Year"].to_numpy()
data_flow = data["Flow"].to_numpy()
data_height = data["Height"].to_numpy()
```

Plot Figures

```python
plt.figure()
plt.scatter(data_year, data_flow, color="blue", label="Flow")
plt.legend()
plt.grid(axis="both")
plt.xlabel("Year")
plt.ylabel("Flow (mÂ³/s)")
plt.show()
```

```python
plt.figure()
plt.scatter(data_year, data_height, color="red", label="Flow")
plt.legend()
plt.grid(axis="both")
plt.xlabel("Year")
plt.ylabel("Height (m)")
plt.show()
```

Determing Missing heights

```python
#calculating coeficients by second order linear regression
data2 = data.dropna(axis='rows')
data2=pd.DataFrame(np.array(data2),columns=['Year','Flow','Height'])

data2_year= data2["Year"].to_numpy()
data2_flow = data2["Flow"].to_numpy()
data2_height = data2["Height"].to_numpy()
```

Histogram Plot

```python
fig, ax = plt.subplots(1,2,figsize=(12,5))

sns.distplot(data_height,ax=ax[0],label='Height')
ax[0].set_yticks([])
ax[0].set_xlabel('Heigh')
ax[0].legend()
sns.distplot(data2_height,ax=ax[1],label='Height')
ax[1].set_yticks([])
ax[1].set_xlabel('Heigh w/ added values')
ax[1].legend()



plt.tight_layout()
# plt.savefig('Fig3.eps',format = 'eps')
plt.show()
```
